```python
import orchest
from matplotlib import pyplot as plt

telemetry_d1, telemetry_d2 = orchest.get_inputs()['telemetry']

fig, ax = plt.subplots(3, figsize=(20,10))
fig.suptitle("Fastest Race Lap Telemetry Comparison")

ax[0].plot(telemetry_d1['Distance'], telemetry_d1['Speed'], label=orchest.get_pipeline_param('d1'))
ax[0].plot(telemetry_d2['Distance'], telemetry_d2['Speed'], label=orchest.get_pipeline_param('d2'))
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")
ax[1].plot(telemetry_d1['Distance'], telemetry_d1['Throttle'], label=orchest.get_pipeline_param('d1'))
ax[1].plot(telemetry_d2['Distance'], telemetry_d2['Throttle'], label=orchest.get_pipeline_param('d2'))
ax[1].set(ylabel='Throttle')
ax[2].plot(telemetry_d1['Distance'], telemetry_d1['Brake'], label=orchest.get_pipeline_param('d1'))
ax[2].plot(telemetry_d2['Distance'], telemetry_d2['Brake'], label=orchest.get_pipeline_param('d2'))
ax[2].set(ylabel='Brakes')
# Hide x labels and tick labels for top plots and y ticks for right plots.
for a in ax.flat:
    a.label_outer()
    
plt.show()
```
